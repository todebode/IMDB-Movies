<b>1. Loading the tsv files with Pandas

```python
basics_url="https://datasets.imdbws.com/title.basics.tsv.gz"
ratings_url="https://datasets.imdbws.com/title.ratings.tsv.gz"
akas_url="https://datasets.imdbws.com/title.akas.tsv.gz"
```

```python
import pandas as pd
basics = pd.read_csv(basics_url,sep='\t', low_memory=False)
ratings = pd.read_csv(ratings_url,sep='\t', low_memory=False)
akas = pd.read_csv(akas_url,sep='\t', low_memory=False)
basics.head()
```

<b>2. Required Preprocessing

            Filtering/Cleaning Steps

<b>Title Basics: 
    
- Replace "\N" with np.nan
    
- Eliminate movies that are null for runtimeMinutes
    
- Eliminate movies that are null for genre
    
- keep only titleType==Movie
    
- keep startYear 2000-2022
    
- Eliminate movies that include  "Documentary" in genre 
    
<b>AKAs:
    
- keep only US entries.
    
- Replace "\N" with np.nan
    
<b>Ratings:
    
- Replace "\N" with np.nan (if any)

                           Cleaning steps for Title Basics

```python
# Replace "\N" with np.nan
basics = basics.replace({'\\N':'np.nan'}) 
basics
```

```python
# Eliminate movies that are null for runtimeMinutes
basics_filter = basics['runtimeMinutes']!='np.nan'
basics = basics[basics_filter]
basics
```

```python
# Eliminate movies that are null for genre
basics_filter1 = basics['genres']!='np.nan'
basics = basics[basics_filter1]
basics
```

```python
# keep only titleType==Movie
basics_filter2 = basics['titleType']=='movie'
basics = basics[basics_filter2]
basics
```

```python
basics.info()
```

```python
# keep startYear 2000-2022
basics_filter3 = basics['startYear']>= '2000'
basics = basics[basics_filter3]
basics
```

```python
# Eliminate movies that include  "Documentary" in genre
basics_filter4 = basics['genres'].str.contains('documentary',case=False)
basics = basics[~basics_filter4]
basics
```

                                    Cleaning steps for AKAs

```python
akas.head()
```

```python
akas['region'].unique()
```

```python
# keep only US entries
akas_filter = akas['region']== 'US'
akas = akas[akas_filter]
akas
```

```python
# Replace "\N" with np.nan
akas = akas.replace({'\\N':'np.nan'}) 
akas
```

```python
# Filter the basics table down to only include the US by using the filter 
# akas dataframe
basics_filter5 =basics['tconst'].isin(akas['titleId'])
basics = basics[~basics_filter5]
basics
```

                                 Cleaning steps for Ratings

```python
ratings.head()
```

```python
# Replace "\N" with np.nan (if any)
ratings = ratings.replace({'\\N':'np.nan'}) 
ratings
```

<b>3. Exporting DataFrames to CSV and saving compressed .csv.gz Files

```python
basics.to_csv('basics_imdb.csv')
ratings.to_csv('ratings_imdb.csv')
akas.to_csv('akas_imdb.csv')
```
